Agente inteligente para atención al cliente de bancos

```python
class AgenteBanco:
    def __init__(self):
        self.percepciones = [
            "Saldos", "Pagos", "Transferencias", "Tarjetas", "Préstamos",
            "Seguridad/Fraude", "Bloqueos", "Promociones", "Movimientos",
            "Configuración", "Chatbot", "Agente_humano", "Estado_de_cuenta",
            "Historial", "Ayuda_rápida", "Cerrar_cuenta", "Actualizar_datos",
            "Horarios", "Cambiar_idioma", "Otra_consulta"
        ]


    def pedir_numero(self, prompt, minimo=None, maximo=None):
        while True:
            entrada = input(prompt).strip()
            if entrada == "":
                print("Debe ingresar un número. No deje vacío.")
                continue
            try:
                valor = int(entrada)
                if minimo is not None and valor < minimo:
                    print(f"Debe ser mínimo {minimo}.")
                    continue
                if maximo is not None and valor > maximo:
                    print(f"Debe ser máximo {maximo}.")
                    continue
                return valor
            except ValueError:
                print("Entrada inválida: solo números enteros.")


    def mostrar_menu_principal(self):
        print("\n MENÚ PRINCIPAL DEL BANCO ")
        print("""
1) Saldos
2) Pagos
3) Transferencias
4) Tarjetas
5) Préstamos
6) Seguridad / Fraude
7) Bloqueos
8) Promociones
9) Movimientos
10) Configuración
11) Chatbot
12) Agente humano
13) Estado de cuenta
14) Historial
15) Ayuda rápida
16) Cerrar cuenta
17) Actualizar datos
18) Horarios
19) Cambiar idioma
20) Otra consulta
0) Salir
""")


    def ejecutar_submenu_generico(self, titulo, opciones):
        print(f"\n SUBMENÚ: {titulo} ")
        for i, texto in enumerate(opciones, 1):
            print(f"{i}) {texto}")
        print("0) Volver")
        op = self.pedir_numero("Seleccione: ", 0, len(opciones))

        if op == 0:
            return


        print("\n Opciones adicionales ")
        print("1) Confirmar proceso")
        print("2) Enviar comprobante")
        print("3) Cancelar")
        print("0) Volver")
        s = self.pedir_numero("Seleccione una opción adicional: ", 0, 3)

        if s == 0:
            return
        if s == 3:
            print("El proceso ha sido cancelado.\n")
            return

        print(" Proceso realizado exitosamente.\n")



    def submenu_saldos(self):
        print("\n SUBMENÚ: SALDOS ")
        print("1) Saldo corriente")
        print("2) Saldo disponible")
        print("3) Por periodos")
        print("0) Volver")
        o = self.pedir_numero("Seleccione: ", 0, 3)
        if o == 0:
            return

        print("\n Método de entrega ")
        print("1) Correo\n2) SMS\n3) PDF\n0) Volver")
        m = self.pedir_numero("Seleccione: ", 0, 3)

        if m == 0:
            return

        print(" Proceso realizado: saldo enviado.\n")

    def submenu_pagos(self):
        print("\n SUBMENÚ: PAGOS ")
        print("1) Servicios públicos")
        print("2) Tarjeta de crédito")
        print("3) Programar pago")
        print("0) Volver")
        o = self.pedir_numero("Seleccione: ", 0, 3)
        if o == 0:
            return

        print("\n Confirmación del pago ")
        print("1) Confirmar")
        print("2) Enviar comprobante")
        print("3) Cancelar")
        print("0) Volver")
        s = self.pedir_numero("Seleccione: ", 0, 3)

        if s == 0:
            return
        if s == 3:
            print("Pago cancelado.\n")
            return

        print(" Proceso realizado: pago completado.\n")

    def submenu_transferencias(self):
        print("\n--- SUBMENÚ: TRANSFERENCIAS ---")
        print("1) Cuentas propias")
        print("2) A terceros")
        print("3) Internacional")
        print("0) Volver")
        o = self.pedir_numero("Seleccione: ", 0, 3)
        if o == 0:
            return

        print("\n-- Confirmación --")
        print("1) Confirmar")
        print("2) Enviar comprobante")
        print("3) Cancelar")
        print("0) Volver")
        s = self.pedir_numero("Seleccione: ", 0, 3)

        if s == 0:
            return
        if s == 3:
            print("Transferencia cancelada.\n")
            return

        print(" Transferencia realizada.\n")




    def submenu_tarjetas(self):
        self.ejecutar_submenu_generico("Tarjetas", [
            "Activar tarjeta",
            "Bloquear tarjeta",
            "Solicitar reposición"
        ])

    def submenu_prestamos(self):
        self.ejecutar_submenu_generico("Préstamos", [
            "Consultar préstamo",
            "Simular nuevo crédito"
        ])

    def submenu_seguridad(self):
        self.ejecutar_submenu_generico("Seguridad / Fraude", [
            "Reportar transacción sospechosa",
            "Activar monitoreo"
        ])

    def submenu_bloqueos(self):
        self.ejecutar_submenu_generico("Bloqueos", [
            "Bloquear producto",
            "Desbloquear producto"
        ])

    def submenu_promociones(self):
        self.ejecutar_submenu_generico("Promociones", [
            "Promociones por tarjeta",
            "Promociones por cuenta"
        ])

    def submenu_movimientos(self):
        self.ejecutar_submenu_generico("Movimientos", [
            "Últimos 5 movimientos",
            "Últimos 30 días"
        ])

    def submenu_configuracion(self):
        self.ejecutar_submenu_generico("Configuración", [
            "Cambiar clave",
            "Actualizar perfil"
        ])

    def submenu_chatbot(self):
        self.ejecutar_submenu_generico("Chatbot", [
            "Chat rápido",
            "Chat con historial"
        ])

    def submenu_agente(self):
        self.ejecutar_submenu_generico("Agente humano", [
            "Llamar agente",
            "Chat en vivo"
        ])

    def submenu_estado_cuenta(self):
        self.ejecutar_submenu_generico("Estado de cuenta", [
            "Generar PDF",
            "Enviar por correo"
        ])

    def submenu_historial(self):
        self.ejecutar_submenu_generico("Historial", [
            "Historial completo",
            "Historial filtrado"
        ])



    def manejar_opcion(self, o):
        if o == 1: self.submenu_saldos()
        elif o == 2: self.submenu_pagos()
        elif o == 3: self.submenu_transferencias()
        elif o == 4: self.submenu_tarjetas()
        elif o == 5: self.submenu_prestamos()
        elif o == 6: self.submenu_seguridad()
        elif o == 7: self.submenu_bloqueos()
        elif o == 8: self.submenu_promociones()
        elif o == 9: self.submenu_movimientos()
        elif o == 10: self.submenu_configuracion()
        elif o == 11: self.submenu_chatbot()
        elif o == 12: self.submenu_agente()
        elif o == 13: self.submenu_estado_cuenta()
        elif o == 14: self.submenu_historial()
        elif o == 15: print(" Ayuda rápida mostrada.\n")
        elif o == 16: print(" Proceso de cierre de cuenta iniciado.\n")
        elif o == 17: print(" Datos actualizados.\n")
        elif o == 18: print(" Horarios mostrados.\n")
        elif o == 19: print(" Idioma cambiado.\n")
        elif o == 20: print(" Consulta registrada.\n")
        elif o == 0:
            print("\nGracias por usar el sistema.")
        else:
            print("Opción inválida.")


    def iniciar(self):
        print("Bienvenido al sistema del banco.\n")
        while True:
            self.mostrar_menu_principal()
            o = self.pedir_numero("Seleccione una opción: ", 0, 20)
            if o == 0:
                self.manejar_opcion(0)
                break
            self.manejar_opcion(o)

```

```python
agente = AgenteBanco()
agente.iniciar()
```
